## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset 

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('EDA_data.csv')
df.head(2)
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df1 = df.copy()
df1.head(1)
```

## creating dummies

```python
df1.columns
```

```python
df2 = df1[[ 'Rating','Size','Type of ownership', 'Sector', 'Revenue','job_state','Age_of_company', 'Industry',
           'python', 'R', 'excel', 'sql', 'spark', 'tableau','aws', 'hadoop', 'Hourly_salary', 'Employ_prov_salary',
           'job_simplified','job_seniority', 'descpn_lngth', 'avg_salary']]
df_dummies = pd.get_dummies(df2)
df_dummies.head(3)
```

```python
df_dummies.shape
```

## splitting dataset 

```python
X = df_dummies.drop("avg_salary", axis = 1)
y = df_dummies.avg_salary.values
print("shape of X = ", X.shape)
print("shape of y = ", y.shape)
```

## train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size = 0.10, random_state=51)
print("shape of X_train = ", X_train.shape)
print("shape of y_train = ", y_train.shape)
print("shape of X_test = ", X_test.shape)
print("shape of y_test = ", y_test.shape)
```

```python
#scaling the features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

## models

```python
from sklearn.metrics import confusion_matrix, r2_score
from sklearn.model_selection import cross_val_score
```

```python
#linear regg
import sklearn.linear_model as linear_model
LR = linear_model.LinearRegression()
LR.fit(X_train, y_train)
y_pred_lr = LR.predict(X_test)
r2_score(y_test, y_pred_lr)
```

```python
cross_val_score(LR, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 3 )
```

```python
cross_val_score(LR, X_train, y_train, scoring = 'r2', cv = 3 )
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred_lr)
```

```python
#ridge 
LRRidge = linear_model.Ridge()
LRRidge.fit(X_train, y_train)
y_pred_lrridge = LRRidge.predict(X_test)
r2_score(y_test, y_pred_lrridge)
```

```python
cross_val_score(LRRidge, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 4 )
```

```python
#lasso
LRLasso = linear_model.Lasso()
LRLasso.fit(X_train, y_train)
y_pred_lrlasso = LRLasso.predict(X_test)
r2_score(y_test, y_pred_lrlasso)
```

```python
cross_val_score(LRLasso, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 3 )
```

```python
#svm
from sklearn.svm import SVR
svr_reg2 = SVR(kernel = 'rbf')
svr_reg2.fit(X_train, y_train)
y_pred_svr2 = svr_reg2.predict(X_test)
r2_score(y_test, y_pred_svr2)
```

```python
svr_reg3 = SVR(kernel = 'rbf')
svr_reg3.fit(X_train_sc, y_train)
y_pred_svr3 = svr_reg3.predict(X_test_sc)
r2_score(y_test, y_pred_svr3)
```

```python
cross_val_score(LRLasso, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 3 )
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators = 140)
rf_reg.fit(X_train, y_train)
y_pred_rf = rf_reg.predict(X_test)
r2_score(y_test, y_pred_rf)
```

```python
cross_val_score(rf_reg, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 3 )
```

```python
cross_val_score(rf_reg, X_train, y_train, scoring = 'r2', cv = 3 )
```

```python
np.mean(cross_val_score(rf_reg, X_train, y_train, scoring = 'r2', cv = 3 ))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred_rf)
```
