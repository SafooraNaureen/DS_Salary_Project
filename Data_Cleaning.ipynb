### salary parsing

```python
df.shape
```

```python
df1 = df.copy()
df1 = df1[df1['Salary Estimate'] != '-1']
df1.shape
```

```python
df1['Hourly_salary'] = df1['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df1[['Hourly_salary','Salary Estimate']].head(100)
```

```python
df1['Employ_prov_salary'] = df1['Salary Estimate'].apply(lambda x: 1 if 'Employer Provided Salary:' in x else 0)
df1[['Employ_prov_salary','Salary Estimate']].head(100)
```

```python
salary = df1['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
salary_minusKd = salary.apply(lambda x: x.replace('K',' ').replace('$',' '))
salary_minusKd
```

```python
salary_minus_hr_emp = salary_minusKd.apply(lambda x: x.lower().replace('per hour', ' ')
                                          .replace('employer provided salary:',' ')
                                          .replace('\n','')
                                          .replace('\r\n','')
                                          .replace('\r',''))

salary_minus_hr_emp
```

```python
df1['min_salary'] = salary_minus_hr_emp.apply(lambda x: int(x.split('-')[0]))
df1[['Salary Estimate','min_salary']].head()
```

```python
df1['max_salary'] = salary_minus_hr_emp.apply(lambda x: int(x.split('-')[1]))
df1[['Salary Estimate','min_salary','max_salary']].head()
```

```python
df1['avg_salary'] = (df1.min_salary+df1.max_salary)/2
df1[['Salary Estimate','min_salary','max_salary','avg_salary']].head()
```

### company name (text only) 

```python
df1['Company_txt'] = df1.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-4], axis = 1)
df1[['Company_txt','Company Name']].head()
```

### state field

```python
df1['Location'] = df1['Location'].apply(lambda x: x.replace('Greenwood Village,',''))
df1['job_state']  = df1['Location'].apply(lambda x: x.split(',')[1])
df1['job_state'].value_counts()
```

### age of company

```python
df1['Age_of_company'] = df1['Founded'].apply(lambda x: x if x<0 else 2020-x)
df1[['Age_of_company','Founded']].head(10)
```

### parsing job description 

```python
df1['python'] = df1['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df1['python'].value_counts()
```
